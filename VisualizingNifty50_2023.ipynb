```python
train_ticker_list = list(nse_stockdata.keys())
dow = {'Monday':0,'Tuesday':0.25,'Wednesday':0.5,'Thursday':0.75,'Friday':1}
def get_sample_stock_data(nse_stockdata,ticker,start_id,end_id,time_series_length=40): #Process the data and Normalize it.

  data = nse_stockdata[ticker]

  fdata = data.iloc[start_id:end_id,:].copy()

  max_vol = fdata['Volume'].max()
  max_close = fdata['Close'].max()
  max_open = fdata['Open'].max()
  max_high = fdata['High'].max()
  max_low = fdata['Low'].max()

  min_vol = fdata['Volume'].min()
  min_close = fdata['Close'].min()
  min_open = fdata['Open'].min()
  min_high = fdata['High'].min()
  min_low = fdata['Low'].min()


  global_price_max = max(max_open,max_close,max_high,max_low)
  global_price_min = min(min_open,min_close,min_high,min_low)

  #Normalize
  fdata['Open'] = (fdata['Open']-global_price_min)/(global_price_max-global_price_min+0.01)
  fdata['High'] = (fdata['High']-global_price_min)/(global_price_max-global_price_min+0.01)
  #print(fdata[['Open','High']].values.shape)
  return fdata[['Open','High']].values
```

```python
stock_times = [] #There will be <lenstamps> lists. Each list contains all 50 NSE stock and each with 40 day vectors
from tqdm import tqdm  # Progress bar
lenstamps = 100
step = 1
with tqdm(total=lenstamps*50, desc="Generating data for nse", unit="timseries") as pbar:
  for i in range(lenstamps):
    start_id = total_length - step*(i+1) -40
    end_id = start_id + 40
    stock_list = []
    for ticker in train_ticker_list:

      data = get_sample_stock_data(nse_stockdata,ticker,start_id,end_id) #this function can be optimized very highly/
      stock_list.append(data)
      pbar.update(1)

    stock_times.append(stock_list)

stock_times = list(reversed(stock_times))
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class TimeSeriesVAE(nn.Module):
    def __init__(self, input_channels=5, input_length=40, latent_dim=20):
        super(TimeSeriesVAE, self).__init__()

        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv1d(input_channels, 16, kernel_size=3, stride=2, padding=1),  # (B, 16, 20)
            nn.ReLU(),
            nn.Conv1d(16, 32, kernel_size=3, stride=1, padding=1),  # (B, 32, 20)
            nn.ReLU(),
            nn.Conv1d(32, 64, kernel_size=3, stride=2, padding=1),  # (B, 64, 10)
            nn.ReLU()
        )

        # Fully connected layers for latent space
        self.flatten = nn.Flatten()
        self.fc_mu = nn.Linear(64 * (input_length // 4), latent_dim)  # Mean of latent space
        self.fc_logvar = nn.Linear(64 * (input_length // 4), latent_dim)  # Log-variance of latent space

        # Fully connected layers for decoding
        self.fc_dec = nn.Linear(latent_dim, 64 * (input_length // 4))

        # Decoder
        self.decoder = nn.Sequential(
            nn.ConvTranspose1d(64, 32, kernel_size=3, stride=2, padding=1, output_padding=1),  # (B, 32, 20)
            nn.ReLU(),
            nn.ConvTranspose1d(32, 16, kernel_size=3, stride=1, padding=1),  # (B, 16, 20)
            nn.ReLU(),
            nn.ConvTranspose1d(16, input_channels, kernel_size=3, stride=2, padding=1, output_padding=1),  # (B, input_channels, 40)
            nn.Sigmoid()  # Assuming normalized input data
        )

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x):
        # Encode
        x = self.encoder(x)  # (B, 64, 10)
        x = self.flatten(x)  # (B, 64 * 10)
        mu = self.fc_mu(x)  # (B, latent_dim)
        logvar = self.fc_logvar(x)  # (B, latent_dim)

        # Reparameterize
        z = self.reparameterize(mu, logvar)  # (B, latent_dim)

        # Decode
        x_dec = self.fc_dec(z)  # (B, 64 * 10)
        x_dec = x_dec.view(-1, 64, 10)  # Reshape back to (B, 64, 10)
        x_dec = self.decoder(x_dec)  # (B, input_channels, 40)

        return x_dec, mu, logvar

# Instantiate the model
#vae = TimeSeriesVAE(input_channels=2, input_length=40, latent_dim=2)


```

#### Visualize 3 stocks, ADANI PORTS, TATAMOTORS and BPCL in latent space as well as time series accross the whole year

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import numpy as np

# Assuming latent_stock_times is a list where each element is a 2D array (n, 2)
# Example: latent_stock_times[0] = np.random.randn(100, 2)
# Assuming stock_times is a list of time series where each element contains a 2D array with shape (n, time_steps)
# Example: stock_times[0] = np.random.randn(50, 40)

# Set up the figure and axis for the latent space plot
fig, (ax_latent, ax_stock) = plt.subplots(1, 2, figsize=(16, 6))

# Define the 3 specific stocks whose points and time series will be plotted
stocks = [1, 43, 10]
stock_names = ["ADANI PORTS", "TATAMOTORS", "BPCL"]

# Generate a fixed color map for the 3 selected stocks
colors = plt.cm.viridis(np.linspace(0, 1, len(stocks)))  # Generate 3 distinct colors from the 'viridis' colormap

# Scatter plot for the latent space
sc = ax_latent.scatter([], [], c=[], cmap='viridis', marker='o')

# Set axis limits for the latent space plot (modify based on your data range)
ax_latent.set_xlim(-2, 2)
ax_latent.set_ylim(-2, 2)
ax_latent.set_xlabel('Latent Dimension 1')
ax_latent.set_ylabel('Latent Dimension 2')
ax_latent.set_title('Latent Space Animation')

# Set axis limits for the time series plot (modify based on your time series range)
ax_stock.set_xlim(0, 40)  # Assuming 40 time steps for each series
ax_stock.set_ylim(-2, 2)  # Modify based on the range of your time series
ax_stock.set_xlabel('Time Steps')
ax_stock.set_ylabel('Amplitude')
ax_stock.set_title('Stock Time Series')

# Initialize function: set the scatter plot and time series plot to an empty state
def init():
    sc.set_offsets(np.empty((0, 2)))  # Start with an empty plot with correct shape
    # Clear the time series plot
    ax_stock.clear()
    ax_stock.set_xlim(0, 40)
    ax_stock.set_ylim(-2, 2)
    ax_stock.set_xlabel('Time Steps')
    ax_stock.set_ylabel('Amplitude')
    ax_stock.set_title('Stock Time Series')
    return sc,

# Update function: update both the latent space scatter plot and the time series plot for each frame
def update(frame):
    # Update latent space scatter plot
    data = latent_stock_times[frame]
    if data.ndim == 1:
        data = data.reshape(-1, 1)  # Reshape to (n, 1)
        if data.shape[1] == 1:
            data = np.repeat(data, 2, axis=1)
    elif data.shape[1] != 2 and data.ndim != 1:
        data = data[:, :2]

    # Select the rows corresponding to the selected stocks
    data = data[stocks]
    sc.set_offsets(data)  # Update latent space data
    sc.set_array(np.linspace(0, 1, len(stocks)))  # Assign color values for each point

    # Update time series plot
    ax_stock.clear()
    ax_stock.set_xlim(0, 40)
    ax_stock.set_ylim(-2, 2)  # Modify based on your data range
    ax_stock.set_xlabel('Time Steps')
    ax_stock.set_ylabel('Amplitude')
    ax_stock.set_title('Stock Time Series')

    # Plot the first channel of each time series for the selected stocks in stock_times[frame]
    for i, stock_idx in enumerate(stocks):  # Loop over the selected stocks
        ax_stock.plot(stock_times[frame][stock_idx][:, 0], label=stock_names[i], color=colors[i])  # Plot with consistent color and name

    # Add legend with stock names
    ax_stock.legend()

    return sc,

# Create the animation using FuncAnimation
ani = FuncAnimation(fig, update, frames=len(latent_stock_times), init_func=init, blit=True)

# To display the animation in Google Colab
HTML(ani.to_html5_video())
```
